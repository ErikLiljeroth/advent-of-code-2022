```python
filename = 'input.txt'

with open(filename, 'r') as f:
    input = f.readlines()
```

```python
input
```

```python
crate_rows = []
crate_nbrs = []
instrc = []
for row in input:
    if row.startswith('move'):
        z = row.replace('move ', '').replace('from ', '').replace('to ', '').strip()
        z = z.split(' ')
        z = [int(x) for x in z]
        instrc.append(z)
    elif row.startswith(' 1'):
        x = row.strip().replace('  ', '')
        x = x.split(' ')
        crate_nbrs = x
    elif row.startswith('\n'):
        pass
    else:
        x = row.replace('    ', ' ')
        x = x.split(' ')
        x = [r.strip('\n') for r in x]
        crate_rows.append(x)

crate_nbrs = [int(c) for c in crate_nbrs]

#letters = row.replace('[', '').replace(']', '').strip()
#letters_list = letters.split(' ')
```

```python
row = input[2]

x = row.replace('    ', ' ')
x = x.split(' ')
x = [r.strip('\n') for r in x]
print(x)
```

```python
# create stacks
stacks = {}
for idx,nbr in enumerate(crate_nbrs):
    cur_stack = []
    for crate_row in crate_rows:
        if crate_row[nbr-1] != '':
            cur_stack.append(crate_row[nbr-1])
    stacks[nbr] = cur_stack
    stacks[nbr].reverse()
```

```python
stacks
```

```python
#instrcts part I
import copy
stacks_adj = copy.copy(stacks)

for ins in instrc:
    for t in range(ins[0]):
        elem = stacks[ins[1]].pop()
        stacks[ins[2]].append(elem)

```

```python
stacks
```

```python
# instrct part II
import copy
stacks_adj = copy.copy(stacks)

# nbr from to
for ins in instrc:
    if ins[0] == 1:
        elem = stacks_adj[ins[1]].pop()
        stacks_adj[ins[2]].append(elem)

    if ins[0] > 1:
        crate_set = []
        for t in range(ins[0]):
            crate_set.append(stacks_adj[ins[1]].pop())
        crate_set.reverse()
        stacks_adj[ins[2]].extend(crate_set)
    else:
        pass

```
